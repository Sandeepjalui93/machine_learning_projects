```python
# Convert Start_Time and End_Time to datetypes
dataset['Start_Time'] = pd.to_datetime(dataset['Start_Time'], errors='coerce')
dataset['End_Time'] = pd.to_datetime(dataset['End_Time'], errors='coerce')

# Extract year, month, day, hour and weekday
dataset['Year']=dataset['Start_Time'].dt.year
dataset['Month']=dataset['Start_Time'].dt.strftime('%b')
dataset['Day']=dataset['Start_Time'].dt.day
dataset['Hour']=dataset['Start_Time'].dt.hour
dataset['Weekday']=dataset['Start_Time'].dt.strftime('%a')

# Extract the amount of time in the unit of minutes for each accident, round to the nearest integer
td='Time_Duration(min)'
dataset[td]=round((dataset['End_Time']-dataset['Start_Time'])/np.timedelta64(1,'m'))
dataset.info()
```

```python
# Check if there is any negative time_duration values
dataset[td][dataset[td]<=0]
```

```python
# Drop the rows with td<0

neg_outliers=dataset[td]<=0

# Set outliers to NAN
dataset[neg_outliers] = np.nan

# Drop rows with negative td
dataset.dropna(subset=[td],axis=0,inplace=True)
dataset.info()
```

```python
# Double check to make sure no more negative td
dataset[td][dataset[td]<=0]
```

```python
# Remove outliers for Time_Duration(min): n * standard_deviation (n=3), backfill with median

n=3

median = dataset[td].median()
std = dataset[td].std()
outliers = (dataset[td] - median).abs() > std*n

# Set outliers to NAN
dataset[outliers] = np.nan

# Fill NAN with median
dataset[td].fillna(median, inplace=True)

dataset.info()
```

```python
# Print time_duration information
print('Max time to clear an accident: {} minutes or {} hours or {} days; Min to clear an accident td: {} minutes.'.format(dataset[td].max(),round(dataset[td].max()/60), round(dataset[td].max()/60/24), dataset[td].min()))
```

# Train and test data split 

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=0)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
independent_scalar = StandardScaler()
X_train_std = independent_scalar.fit_transform (X_train) #fit and transform
X_test_std = independent_scalar.transform (X_test) # only transform
```

# Classification model:- 1) Logistic Regression

```python
# import the class
from sklearn.linear_model import LogisticRegression

# instantiate the model (using the default parameters)
logreg = LogisticRegression(max_iter=1000)

# fit the model with data
logreg.fit(X_train,y_train)
```

# Make Predictions

```python
pred_log = logreg.predict(X_test)
```

# Evaluating the Model

```python
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, pred_log)
cnf_matrix
```

```python
class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
# Confusion Matrix Evaluation Metrics
print("Accuracy:",metrics.accuracy_score(y_test, pred_log))
```
